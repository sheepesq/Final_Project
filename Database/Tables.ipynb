```python
import pandas as pd
import psycopg2
import sqlalchemy
from sqlalchemy import create_engine
```

```python
url = "https://bootcamp-citibike-bucket.s3.amazonaws.com/NYC/May2016.csv"
df2016 = pd.read_csv(url, sep=',', header= 0)
df2016.head(10)
```

```python
station_id = []

for ind in df2016.index:
    if df2016['start station id'][ind] not in station_id:
        station_id.append(df2016['start station id'][ind])
        
station_id.sort()
print(station_id)
```

```python
station_name = []
name = ""

for i in range(len(station_id)):
    name = df2016.loc[(station_id[i]), "start station name"]
    station_name.append(name)

print(station_name)
```

```python
latitude = []
lat = ""

for i in range(len(station_id)):
    lat = df2016.loc[(station_id[i]), "start station latitude"]
    latitude.append(lat)

print(latitude)
```

```python
longitude = []
long = ""

for i in range(len(station_id)):
    long = df2016.loc[(station_id[i]), "start station longitude"]
    longitude.append(long)

print(longitude)
```

```python
data = {"Station ID": station_id,
        "Station Name": station_name,
        "Station Latitude": latitude,
        "Station Longitude": longitude}

stationdf2016 = pd.DataFrame(data)
```

```python
stationdf2016.head()
```
