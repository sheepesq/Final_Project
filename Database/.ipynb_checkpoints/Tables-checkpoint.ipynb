```python
import pandas as pd
import psycopg2
import sqlalchemy
from sqlalchemy import create_engine
```

```python
url = "https://bootcamp-citibike-bucket.s3.amazonaws.com/NYC/May2016.csv"
df2016 = pd.read_csv(url, sep=',', header= 0)
df2016.head(10)
```

```python
df2016.drop_duplicates(subset = "start station id", inplace=True)

df2016
```

```python
df2016 = df2016.drop(["tripduration", "starttime", "stoptime", "end station id", "end station name", "end station latitude",
            "end station longitude", "bikeid", "usertype", "birth year", "gender"], axis=1)

df2016
```

```python
df2016 = df2016.rename(columns={'start station id': 'station id', 'start station name': 'station name',
                               'start station latitude': 'latitude', 'start station longitude': 'longitude'})
df2016
```
