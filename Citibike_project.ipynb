```python
# Import dependencies
import pandas as pd
```

```python
# Import our input dataset
citi_May21_df = pd.read_csv('Resources/JC-202105-citibike-tripdata.csv', index_col=0)
citi_May21_df
```

```python
### Explore data
start = citi_May21_df['start_station_name'].unique()
end = citi_May21_df['end_station_name'].unique()
```

```python
start
```

```python
# Identify ending locations that were never starting locations
not_start = []
for loc in end:
    if loc not in start:
        not_start.append(loc)
    
not_start
```

```python
# Identify any starting locations that were never ending locations
not_end = []
for loc in start:
    if loc not in end:
        not_end.append(loc)
    
not_end
```
