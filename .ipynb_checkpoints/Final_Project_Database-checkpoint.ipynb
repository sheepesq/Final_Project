```python
import pandas as pd
import requests
import psycopg2
from sqlalchemy import create_engine
```

```python
url = "https://bootcamp-citibike-bucket.s3.amazonaws.com/NYC/May2016.csv"
df2016 = pd.read_csv(url, sep=',', header= 0)
df2016.head(10)
```

```python
engine = create_engine('postgresql://postgres:postgres@citibike-analysis.cbk1sw5fz5td.us-east-1.rds.amazonaws.com:5432/project')
df2016.to_sql('may2016', engine)
```
