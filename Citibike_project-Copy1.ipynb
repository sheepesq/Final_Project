```python
# Import dependencies
import pandas as pd
import psycopg2
```

```python
# This code is adapted from the tutorial hosted below:
# http://www.postgresqltutorial.com/postgresql-python/connect/


# Establish a connection to the database by creating a cursor object
# The PostgreSQL server must be accessed through the PostgreSQL APP or Terminal Shell

host = "citibike-analysis.cbk1sw5fz5td.us-east-1.rds.amazonaws.com"
port = "5432"
database = "project"
user = "postgres" 
password = "postgres"

# Connect to the PostgreSQL database
conn = psycopg2.connect(host=host, port=port, database=database, user=user, password=password)

# Create a cursor object
cur = conn.cursor()

# A function that takes in a PostgreSQL query and outputs a pandas database 
def create_pandas_table(sql_query, database = conn):
    table = pd.read_sql_query(sql_query, database, index_col=['index'])
    return table

# Utilize the create_pandas_table function to create a Pandas data frame for each dataset
may2016_df = create_pandas_table("SELECT * FROM may2016")
may2017_df = create_pandas_table("SELECT * FROM may2017")
may2018_df = create_pandas_table("SELECT * FROM may2018")
may2019_df = create_pandas_table("SELECT * FROM may2019")
stations2016_df = create_pandas_table("SELECT * FROM stations2016")
stations2017_df = create_pandas_table("SELECT * FROM stations2017")
stations2018_df = create_pandas_table("SELECT * FROM stations2018")
stations2019_df = create_pandas_table("SELECT * FROM stations2019")

allmay_df = create_pandas_table("SELECT * FROM allmay")
allstations_df = create_pandas_table("SELECT * FROM allstations")


# Close the cursor and connection to so the server can allocate bandwidth to other requests
cur.close()
conn.close()
```

```python
# Find null values
for column in allmay_df.columns:
    print(f"Column {column} has {allmay_df[column].isnull().sum()} null values")
```

```python
# Drop null rows
allmay_df = allmay_df.dropna()
```

```python
# Remove the stations name Columns (since we have station id columns in place)
allmay_df.drop(columns=["start station name", "end station name"], inplace=True)
allmay_df.head()
```

```python
#Transform string usertype column (Subscriber = 1, Customer = 0)
def change_string(member):
    if member == "Subscriber":
        return 1
    else:
        return 0
    
allmay_df["usertype"] = allmay_df["usertype"].apply(change_string)
allmay_df.head()
```

```python
#  Convert date columns to datetime from object
allmay_df['starttime'] = pd.to_datetime(allmay_df['starttime'])
allmay_df.dtypes
```
